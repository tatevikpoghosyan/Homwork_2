```python
def title(url):
    response=requests.get(url)
    page=TextResponse(url=response.url,body=response.text,encoding="utf-8")
    titles= page.css('article.product_pod > h3 > a::attr(title)').extract()
    return titles
```

```python
all_titles=[]
i=1
while True:
    url='http://books.toscrape.com/catalogue/page-{}.html'.format(i)
    current_titles=title(url)
    if len(current_titles)==0:
        break
    else:
        all_titles.extend(current_titles)
    i=i+1
```

```python
#Price css
```

```python
def price(url):
    response=requests.get(url)
    page=TextResponse(url=response.url,body=response.text,encoding="utf-8")
    prices=[float(i[2:]) for  i in page.css("[class='price_color']::text").extract()]
    return prices
```

```python
all_prices=[]
i=1
while True:
    url='http://books.toscrape.com/catalogue/page-{}.html'.format(i)
    current_prices=price(url)
    if len(current_prices)==0:
        break
    else:
        all_prices.extend(current_prices)
    i=i+1
```

```python
#Rate css
```

```python
def book_rates(url):
    response=requests.get(url)
    page=TextResponse(url=response.url,body=response.text,encoding="utf-8")
    rates = [i[11:].strip() for i in page.css("div[class='image_container']+p::attr(class)").extract()]
    return rates
```

```python
all_rates =[]
i=1
while True:
    url='http://books.toscrape.com/catalogue/page-{}.html'.format(i)
    current_rates=book_rates(url)
    if len(current_rates)==0:
        break
    else:
        all_rates.extend(current_rates)
    i=i+1
```

```python
#Rate xpath
```

```python
all_rates_xpath = []
for i in range(1,51):
    url = 'http://books.toscrape.com/catalogue/page-{}.html'.format(i)
    rates_xpath_fun = [i[11:].strip() for i in page.xpath('//p/@class').extract()[::3]]
    all_rates_xpath.extend(rates_xpath_fun)
print(all_rates_xpath)
```

```python
#Most popular rating is One star 
```

```python
Ratings_with_num=[]
for i in all_rates:
    if i=="One":
        Ratings_with_num.append("1")
    elif i=="Two":
        Ratings_with_num.append("2")
    elif i=="Three":
        Ratings_with_num.append("3")
    elif i=="Four":
        Ratings_with_num.append("4")
    else:
        Ratings_with_num.append("5")
print(Ratings_with_num)
```
